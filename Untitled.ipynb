# McDonald's menu nutrition EDA 




### Background

McDonald's is one the world's most popular fast food chain, spanning across 114 countries with 37,376 outlets. Every menu item on the US menu including breakfast, beef burgers, chicken and fish sandwiches, fries, salads, drinks, coffee and tea, milkshakes, and desserts, has its nutrition information (as of 2017) analyzed in this dataset.


```python
#Importing essesntial tools 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import plotly.express as px
import seaborn as sns
```

```python
df = pd.read_csv('menu.csv')
```

```python
max_values = df[['Serving Size','Calories','Total Fat','Carbohydrates','Dietary Fiber','Sugars','Protein','Vitamin A (% Daily Value)','Vitamin C (% Daily Value)','Calcium (% Daily Value)','Iron (% Daily Value)']].max()
```

According to the dataset, the maximum value for calories and some other variables seems to be high. We will try to figure out which menu items these values correspond to but first we will check to see if infact there is any correlation between calories and other independent variables by plotting a correlation matrix next.

```python
f = plt.figure(figsize=(19,15))
plt.matshow(df.corr(),fignum=f.number)
plt.xticks(range(df.select_dtypes(['number']).shape[1]), df.select_dtypes(['number']).columns, fontsize=14, rotation=45)
plt.yticks(range(df.select_dtypes(['number']).shape[1]), df.select_dtypes(['number']).columns, fontsize=14, rotation=45)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=14)
plt.title('Correlation Matrix', fontsize=16)
```

* The correlation coefficients are proportionate to the box's size and colour intensity.
* As we can see, the variable Calories has strong obvious correlations of the other independent variables whether they are positive or negative.

```python
lists = ["Chicken & Fish","Breakfast","Smoothies & Shakes","Beef & Pork","Snacks & Sides","Desserts","Beverages"]
```

```python
for list in lists:
    items = df.query('Category == list').sort_values(by='Calories',ascending=False)
    items_custom = items.iloc[:,[1,3,5]]
    
    items_t=df.query('Category == list').sort_values(by="Calories", ascending=True)
    items_custom_t=items_t.iloc[:,[1,3,5]]
    
    print(lists.index(list)+1,".",list)
    
    p_express = px.scatter(items_custom, x="Total Fat",y='Calories', color="Item",size='Calories',template="plotly_dark",hover_name="Item")
    p_express.show()
    
    table = items_custom.iloc[:,[0,1]].sort_values(by='Calories',ascending=False).style.hide_index()
    display(table)
    
    fig,ax=plt.subplots()
    x = items_custom_t.iloc[:,0]
    y = items_custom_t.iloc[:,1]
    ax.barh(x,y, tick_label=x)
    plt.show()
```

## Conclusion 

* According to FDA, the recommended adult limit for sodium intake should be less than 2,300 milligrams/day - that's equal to about 1 teaspoon of table salt.

* FDA also recommends adults consume 50 grams of protein per day.

* USDA recommends 66-97 grams of fat per day in a 2000 and 2500 calorie diet respectively.

* NHS recommends daily calorie intake of 2000 kcal for women and 2500 kcal for men.
